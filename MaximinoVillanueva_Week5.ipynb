```python
# ============================================================
# Maximino Villanueva
# Week 5 Python Version
# Tree Based Classification Models Using mlc_churn
# Google Colab Notebook
# ============================================================

# Install required packages.
!pip -q install rpy2 pandas numpy scikit-learn matplotlib seaborn

# Import required Python libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import tools for loading the R modeldata package into Python.
import rpy2.robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.conversion import localconverter

# Import scikit-learn preprocessing and modeling tools.
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

# Import tree based classification models.
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, GradientBoostingClassifier

# Import evaluation tools.
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score, ConfusionMatrixDisplay

# Install and load the R modeldata package inside Colab.
ro.r('if (!requireNamespace("modeldata", quietly = TRUE)) install.packages("modeldata", repos = "https://cloud.r-project.org")')
ro.r('library(modeldata)')
ro.r('data(mlc_churn)')

# Convert the R mlc_churn dataset into a pandas DataFrame.
with localconverter(ro.default_converter + pandas2ri.converter):
    churn_data = ro.conversion.rpy2py(ro.r('mlc_churn'))

# Display the first few rows of the dataset.
display(churn_data.head())

# Show the dataset shape.
print("Dataset shape:", churn_data.shape)

# Remove optional variables listed in the assignment instructions.
columns_to_remove = ["state", "area_code", "account_length", "total_night_calls", "total_eve_calls"]

# Drop optional columns only if they exist in the dataset.
churn_data = churn_data.drop(columns=[col for col in columns_to_remove if col in churn_data.columns])

# Convert the response variable to lowercase text values.
churn_data["churn"] = churn_data["churn"].astype(str).str.lower()

# Separate predictors from the response variable.
X = churn_data.drop(columns=["churn"])

# Convert churn outcome to numeric values where yes equals 1 and no equals 0.
y = churn_data["churn"].map({"no": 0, "yes": 1})

# Identify categorical predictor columns.
categorical_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()

# Identify numeric predictor columns.
numeric_cols = X.select_dtypes(include=["int64", "float64", "int32", "float32"]).columns.tolist()

# Print the categorical columns.
print("Categorical columns:", categorical_cols)

# Print the numeric columns.
print("Numeric columns:", numeric_cols)

# Split data into training and testing sets using stratification.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=123,
    stratify=y
)

# Create numeric preprocessing steps.
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median"))
    ]
)

# Create categorical preprocessing steps.
categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ]
)

# Combine numeric and categorical preprocessing.
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_cols),
        ("cat", categorical_transformer, categorical_cols)
    ]
)

# Create a function to build a bagging classifier that works across sklearn versions.
def make_bagging_model():
    # Try the newer sklearn syntax first.
    try:
        return BaggingClassifier(
            estimator=DecisionTreeClassifier(random_state=123),
            n_estimators=100,
            random_state=123
        )
    # Use older sklearn syntax if needed.
    except TypeError:
        return BaggingClassifier(
            base_estimator=DecisionTreeClassifier(random_state=123),
            n_estimators=100,
            random_state=123
        )

# Create all required models.
models = {
    "CART Decision Tree": DecisionTreeClassifier(random_state=123),
    "Bagged Trees": make_bagging_model(),
    "Random Forest": RandomForestClassifier(n_estimators=300, random_state=123),
    "Boosting Tree Model": GradientBoostingClassifier(random_state=123)
}

# Create empty dictionaries to store model results.
trained_models = {}
results = {}
roc_data = {}

# Loop through each model.
for model_name, model in models.items():
    # Create a modeling pipeline.
    clf = Pipeline(
        steps=[
            ("preprocessor", preprocessor),
            ("model", model)
        ]
    )

    # Fit the model on the training data.
    clf.fit(X_train, y_train)

    # Store the trained model.
    trained_models[model_name] = clf

    # Predict classes on the test data.
    y_pred = clf.predict(X_test)

    # Predict probabilities for the positive churn class.
    y_prob = clf.predict_proba(X_test)[:, 1]

    # Calculate the confusion matrix.
    cm = confusion_matrix(y_test, y_pred)

    # Calculate accuracy.
    accuracy = accuracy_score(y_test, y_pred)

    # Calculate AUC.
    auc = roc_auc_score(y_test, y_prob)

    # Store model results.
    results[model_name] = {
        "Accuracy": accuracy,
        "AUC": auc,
        "Confusion Matrix": cm
    }

    # Calculate ROC curve values.
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)

    # Store ROC curve values.
    roc_data[model_name] = {
        "FPR": fpr,
        "TPR": tpr,
        "AUC": auc
    }

    # Print model results.
    print("\n============================================================")
    print(model_name)
    print("============================================================")
    print("Accuracy:", round(accuracy, 4))
    print("AUC:", round(auc, 4))
    print("\nConfusion Matrix:")
    print(cm)
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=["no", "yes"]))

    # Plot confusion matrix.
    ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["no", "yes"]).plot()
    plt.title(model_name + " Confusion Matrix")
    plt.show()

    # Plot ROC curve.
    plt.figure()
    plt.plot(fpr, tpr, label=model_name + " AUC = " + str(round(auc, 4)))
    plt.plot([0, 1], [0, 1], linestyle=":")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(model_name + " ROC Curve")
    plt.legend()
    plt.show()

# Create a model comparison table.
comparison_table = pd.DataFrame(
    {
        "Model": list(results.keys()),
        "Accuracy": [results[m]["Accuracy"] for m in results.keys()],
        "AUC": [results[m]["AUC"] for m in results.keys()]
    }
)

# Sort the comparison table by AUC.
comparison_table = comparison_table.sort_values(by="AUC", ascending=False)

# Display the model comparison table.
display(comparison_table)

# Plot all ROC curves together.
plt.figure(figsize=(8, 6))

# Add each model ROC curve to one plot.
for model_name, values in roc_data.items():
    plt.plot(values["FPR"], values["TPR"], label=model_name + " AUC = " + str(round(values["AUC"], 4)))

# Add random guessing reference line.
plt.plot([0, 1], [0, 1], linestyle=":")

# Label the x axis.
plt.xlabel("False Positive Rate")

# Label the y axis.
plt.ylabel("True Positive Rate")

# Add the title.
plt.title("ROC Curve Comparison for Tree Based Classification Models")

# Add the legend.
plt.legend()

# Show the plot.
plt.show()

# Function to extract feature names after preprocessing.
def get_feature_names_from_pipeline(pipeline):
    # Get the fitted preprocessor.
    fitted_preprocessor = pipeline.named_steps["preprocessor"]

    # Get numeric feature names.
    numeric_features = numeric_cols

    # Get categorical feature names after one hot encoding.
    categorical_features = fitted_preprocessor.named_transformers_["cat"].named_steps["onehot"].get_feature_names_out(categorical_cols).tolist()

    # Combine all feature names.
    return numeric_features + categorical_features

# Function to extract feature importance.
def get_feature_importance(model_name, pipeline):
    # Get fitted model.
    fitted_model = pipeline.named_steps["model"]

    # Get feature names.
    feature_names = get_feature_names_from_pipeline(pipeline)

    # Handle models with direct feature importance.
    if hasattr(fitted_model, "feature_importances_"):
        importances = fitted_model.feature_importances_

    # Handle bagging model by averaging tree importances.
    elif hasattr(fitted_model, "estimators_"):
        importances = np.mean(
            [tree.feature_importances_ for tree in fitted_model.estimators_ if hasattr(tree, "feature_importances_")],
            axis=0
        )

    # Return None if importance is unavailable.
    else:
        return None

    # Create feature importance table.
    importance_df = pd.DataFrame(
        {
            "Feature": feature_names,
            "Importance": importances
        }
    )

    # Sort by importance.
    importance_df = importance_df.sort_values(by="Importance", ascending=False)

    # Return feature importance table.
    return importance_df

# Loop through models and print top variable importance.
for model_name, pipeline in trained_models.items():
    # Get importance table.
    importance_df = get_feature_importance(model_name, pipeline)

    # Print model name.
    print("\n============================================================")
    print(model_name + " Variable Importance")
    print("============================================================")

    # Display the top 10 variables.
    display(importance_df.head(10))

    # Plot the top 10 variables.
    importance_df.head(10).plot(kind="barh", x="Feature", y="Importance", legend=False)
    plt.title(model_name + " Top 10 Variable Importance")
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    plt.gca().invert_yaxis()
    plt.show()

# Visualize the CART decision tree.
cart_pipeline = trained_models["CART Decision Tree"]

# Get the fitted CART model.
cart_tree = cart_pipeline.named_steps["model"]

# Get feature names after preprocessing.
cart_feature_names = get_feature_names_from_pipeline(cart_pipeline)

# Create a large figure for the tree.
plt.figure(figsize=(24, 12))

# Plot the CART tree.
plot_tree(
    cart_tree,
    feature_names=cart_feature_names,
    class_names=["no", "yes"],
    filled=True,
    rounded=True,
    max_depth=3
)

# Add the title.
plt.title("CART Classification Tree Visualization")

# Show the tree.
plt.show()

# Print final interpretation.
print("\nInterpretation")
print("This notebook compared four tree based classification models on the mlc_churn dataset.")
print("The models included a CART decision tree, bagged trees, random forest, and a boosting tree model.")
print("The Python boosting tree model is used as a practical Python substitute for the C5.0 style tree based model requested in the R instructions.")
print("Each model was evaluated using a confusion matrix, classification report, ROC curve, AUC, and variable importance.")
print("The best model should be selected by comparing test set performance, especially AUC, accuracy, sensitivity, and specificity.")
print("Single decision trees are easier to interpret, while bagging, random forest, and boosting usually improve performance by combining many trees.")
```
